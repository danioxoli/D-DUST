```python
from scipy.interpolate import griddata
import matplotlib.pyplot as plt
import numpy as np
rng = np.random.default_rng()
x = rng.random(10) - 0.5
y = rng.random(10) - 0.5
z = rng.random(10)
display(x, y ,z)
```

```python
X = np.linspace(min(x), max(x))
Y = np.linspace(min(y), max(y))
display(X, Y)
```

```python
X, Y = np.meshgrid(X, Y) # 2D grid for interpolation

grid_z1 = griddata(list(zip(x, y)), z, (X, Y), method='nearest')
grid_z2 = griddata(list(zip(x, y)), z, (X, Y), method='cubic')
display(grid_z1)
display(grid_z2)
```

```python
plt.pcolormesh(X, Y, grid_z1, shading='auto')
plt.plot(x, y, "ok", label="input point")
plt.legend()
plt.colorbar()
plt.axis("equal")
plt.show()

plt.pcolormesh(X, Y, grid_z2, shading='auto')
plt.plot(x, y, "ok", label="input point")
plt.legend()
plt.colorbar()
plt.axis("equal")
plt.show()
```
